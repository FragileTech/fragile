```python
import torch
import pandas as pd
from fragile.benchmarks import Rastrigin, Sphere
from fragile.montezuma import FractalTree

def run(n_walkers):
    dim = 3
    benchmark = Rastrigin(dim)
    device = "cuda"
    fai = FractalTree(n_walkers, env=benchmark, policy=lambda x: torch.randn_like(x) * 2.5)
    state = benchmark.sample(n_walkers).to(device)
    state[:] = state[0, :]
    return fai, state, torch.randn(n_walkers, dim, device=device)
    fai.causal_cone(state, torch.randn(n_walkers, dim))

from holoviews.streams import Pipe
from fragile.dataviz import to_graph
import hvplot
from hvplot.networkx import draw_networkx

def  draw_benchmark(benchmark):
    x, y = benchmark.best_state[:2]
    x_low, y_low = benchmark.bounds.low[:2].numpy(force=True).tolist()
    x_high, y_high = benchmark.bounds.high[:2].numpy(force=True).tolist()
    return hv.Scatter(([x], [y])).opts(marker="star", color="red", size=8, xlim=(x_low, x_high), ylim=(y_low, y_high), alpha=0.7)

def plot_graph(data):
    if not data:
        return hv.Graph([], [], title="Iteration 0")
    fai, i = data
    graph, pos = to_graph(fai)
    return draw_networkx(graph, pos=pos, node_size=40,
                         edge_line_width=1,
                         node_color='reward',
                         edge_color='black',
                         colorbar=True, tools=['hover']).opts(framewise=True, title=f"Iteration {i}")
from time import sleep
def run_fai(observ, action, n_steps, stream, summary_stream):
    
    #fai.reset(observ=observ, action=action)
    df = pd.DataFrame(fai.to_dict())
    summary = pd.DataFrame(fai.summary(), index=[0])
    stream.send(df)
    summary_stream.send(summary)
    for i in range(n_steps):
        fai.step_tree()
        if i % 100 ==0: #fai.will_clone.sum().cpu().item() > 0:
            df = pd.DataFrame(fai.to_dict())
            summary = pd.DataFrame(fai.summary(), index=["value"]).T.reset_index()
            stream.send(df)
            summary_stream.send(summary)
            print(fai.reward.min().cpu().item())
            sleep(0.01)
```

```python
import time
import panel as pn
import param


class Control(param.Parameterized):
    is_running = param.Boolean(default=False)

    def __init__(self, fai, n_steps):
        super().__init__()
        self.reset_btn = pn.widgets.Button(icon="restore", button_type="primary")
        self.play_btn = pn.widgets.Button(icon="player-play", button_type="primary")
        self.pause_btn = pn.widgets.Button(icon="player-pause", button_type="primary")
        self.progress = pn.indicators.Progress(name='Progress', value=0, width=250, max=n_steps)
        self.sleep_val = pn.widgets.FloatInput(value=0.05, width=60)
        self.fai = fai
        self.n_steps = n_steps
        self.curr_step = 0


    def reset(self, event):
        self.fai.reset()
        self.curr_step = 0
        self.progress.value = 0
        self.curr_step = 0
        
    @param.depends("play_btn.clicks")
    def on_play_click(self):
        self.play_btn.disabled = True
        self.pause_btn.disabled = False
        self.is_running = True
        
    @param.depends("pause_btn.value")
    def on_pause_click(self):
        self.play_btn.disabled = False
        self.pause_btn.disabled = True
        self.is_running = False
        

    def run(self):

        if self.is_running:
            self.fai.step_tree()
            self.curr_step += 1
            self.progress.value = self.curr_step
            if self.curr_step >= self.n_steps:
                self.is_running = False
            time.sleep(self.sleep_val.value)
        
        
    def __panel__(self):
        pn.state.add_periodic_callback(self.run, period=1)
        return pn.Column(self.progress, pn.Row(self.play_btn,
                                               self.pause_btn,
                                               self.reset_btn,
                                               pn.pane.Markdown("**Sleep**"),
                                               self.sleep_val),
                         self.on_play_click, self.on_pause_click, self.run)

```

```python
def get_segments(parent, x, y):
    return x[parent], y[parent], x, y

def plot_edges(df, **kwargs):
    parent, x, y = df["parent"].values, df["x"].values, df["y"].values
    segs = x[parent], y[parent], x, y
    kwargs["line_color"] = kwargs.get("line_color", "black")
    return hv.Segments(segs).opts(**kwargs)

def plot_nodes(df, **kwargs):
    return hv.Scatter(df, kdims=["x"], vdims=[ "y"]).opts(**kwargs)

def plot_graph(df, edge_kwargs=None, **node_kwargs):
    edge_kwargs = edge_kwargs or {}
    return plot_edges(df, **edge_kwargs) * plot_nodes(df, **node_kwargs)
```
