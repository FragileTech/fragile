```python
import numpy as np
import networkx as nx
adj = np.array([[0, 1, 1, 0],
       [1, 0, 1, 1],
       [1, 1, 0, 0],
       [0, 1, 0, 0]])
adj = np.random.random((20,20))
def reduce_connections(adj):
    g = nx.Graph(adj)
    mst = nx.minimum_spanning_tree(g)
    paths = nx.shortest_path(mst)
    adj = nx.adjacency_matrix(mst).todense()
    return adj, paths

%timeit adjs, paths = reduce_connections(adj)
```
