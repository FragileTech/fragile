# Emergent Manifold Visualization: Spacetime Bending in the Geometric Gas

This notebook demonstrates the 3D manifold visualization tools for exploring the emergent Riemannian geometry in the Geometric Gas framework.

**Key Concepts:**
- The adaptive diffusion tensor $\Sigma_{\text{reg}}(x,S) = (H + \epsilon_\Sigma I)^{-1/2}$ induces a Riemannian metric $g(x,S) = H + \epsilon_\Sigma I$ on state space
- This metric encodes the fitness landscape's curvature structure
- For **2D runs**: We can visualize this abstract geometry by embedding into 3D space, where the z-axis represents curvature - revealing the "bending of spacetime"
- For **3D+ runs**: We visualize the first 3 spatial dimensions directly

**Reference:** `docs/source/2_geometric_gas/18_emergent_geometry.md`

## Part 2: Standalone Visualization (Using Existing RunHistory)

You can also visualize an existing `RunHistory` object directly, without the configuration dashboard.

```python
# Run a simulation manually
from fragile.core.euclidean_gas import EuclideanGas, CloningParams
from fragile.core.kinetic_operator import KineticOperator, LangevinParams
from fragile.core.companion_selection import CompanionSelection
from fragile.core.fitness import FitnessOperator, FitnessParams
from fragile.bounds import TorchBounds

# Setup
N, d = 1000, 2
bounds = TorchBounds(
    low=torch.full((d,), -6.0),
    high=torch.full((d,), 6.0)
)

companion_selection = CompanionSelection(
    method="uniform",
    epsilon=0.5,
    lambda_alg=0.0,
)

# Create kinetic operator with anisotropic diffusion
kinetic_op = KineticOperator(
    gamma=1.0,
    beta=1.0,
    delta_t=0.05,
    integrator="baoab",
    epsilon_F=0.1,
    use_fitness_force=True,
    use_potential_force=False,
    epsilon_Sigma=0.1,
    use_anisotropic_diffusion=True,  # KEY: Enable geometry
    diagonal_diffusion=False,  # Use full Hessian for complete geometry
    potential=potential_2d,
    device=torch.device("cpu"),
    dtype=torch.float32,
)

fitness_op = FitnessOperator(
    params=FitnessParams(
        alpha=0.7,
        beta=1.3,
        eta=0.1,
        lambda_alg=0.5,
        sigma_min=1e-8,
        A=2.0,
    ),
    companion_selection=companion_selection,
)

# Create cloning parameters
cloning_params = CloningParams(
    sigma_x=0.15,
    lambda_alg=0.0,
    alpha_restitution=0.6,
    alpha=0.7,
    beta=1.3,
    eta=0.1,
    A=2.0,
    sigma_min=1e-8,
    p_max=1.0,
    epsilon_clone=0.005,
    companion_selection=companion_selection,
)

# Create Gas
gas_2d = EuclideanGas(
    N=N,
    d=d,
    companion_selection=companion_selection,
    potential=potential_2d,
    kinetic_op=kinetic_op,
    cloning=cloning_params,
    fitness_op=fitness_op,
    bounds=bounds,
    device=torch.device("cpu"),
    dtype="float32",
    enable_cloning=True,
    enable_kinetic=True,
)

# Initialize walkers
x_init = torch.randn(N, d) * 0.5 + 4.5
x_init = torch.clamp(x_init, min=-6.0, max=6.0)
v_init = torch.randn(N, d) * 0.1

# Run simulation
print("Running 2D simulation with anisotropic diffusion...")
history_2d = gas_2d.run(n_steps=200, x_init=x_init, v_init=v_init)
print(history_2d.summary())
```

```python
# Create standalone visualization
from fragile.experiments.manifold import create_manifold_dashboard

explorer_2d, panel_2d = create_manifold_dashboard(
    history=history_2d,
    fitness_op=fitness_op,  # Provide for on-the-fly Hessian computation
    # Visualization parameters
    show_trajectories=False,
    show_manifold=False,
    show_principal_directions=True,
    epsilon_Sigma=1.,
    curvature_scale=1.,  # Amplify z-displacement for clarity
    embedding_mode="curvature_height",  # Use trace of metric
)

# Display
panel_2d
```

### Exploring Different Embedding Modes

The 2D embedding supports three modes for mapping curvature to z-height:

1. **`curvature_height`** (default): Uses trace of metric $\text{tr}(g) = \sum_i \lambda_i$ 
   - Shows total curvature (sum of all eigenvalues)
   - Good for overall geometry overview

2. **`metric_trace`**: Same as curvature_height (kept for clarity)

3. **`first_eigenvalue`**: Uses largest eigenvalue $\lambda_{\max}(g)$
   - Shows most curved direction
   - Emphasizes strong curvature regions

Try changing the `embedding_mode` parameter in the controls!

## Part 4: Comparing Isotropic vs Anisotropic Diffusion

Let's compare runs with and without anisotropic diffusion to see the geometric effects.

```python
# Helper function to run simulation with specific settings
def run_gas_2d(use_anisotropic: bool, epsilon_Sigma: float = 0.2):
    """Run 2D Gas simulation with configurable diffusion."""
    kinetic_op = KineticOperator(
        gamma=1.0,
        beta=1.0,
        delta_t=0.05,
        integrator="baoab",
        use_potential_force=False,
        epsilon_Sigma=epsilon_Sigma,
        use_anisotropic_diffusion=use_anisotropic,
        diagonal_diffusion=False,
        potential=potential_2d,
        device=torch.device("cpu"),
        dtype=torch.float32,
    )
    
    # Create cloning parameters
    cloning_params = CloningParams(
        sigma_x=0.15,
        lambda_alg=0.5,
        alpha_restitution=0.6,
        alpha=0.7,
        beta=1.3,
        eta=0.1,
        A=2.0,
        sigma_min=1e-8,
        p_max=1.0,
        epsilon_clone=0.005,
        companion_selection=companion_selection,
    )
    
    gas = EuclideanGas(
        N=80,
        d=2,
        companion_selection=companion_selection,
        potential=potential_2d,
        kinetic_op=kinetic_op,
        cloning=cloning_params,
        fitness_op=fitness_op,
        bounds=bounds,
        device=torch.device("cpu"),
        dtype="float32",
    )
    
    x_init = torch.randn(80, 2) * 0.5 + 4.5
    x_init = torch.clamp(x_init, min=-6.0, max=6.0)
    v_init = torch.randn(80, 2) * 0.1
    
    return gas.run(n_steps=150, x_init=x_init, v_init=v_init)

# Run both
print("Running isotropic diffusion...")
history_iso = run_gas_2d(use_anisotropic=False)

print("Running anisotropic diffusion...")
history_aniso = run_gas_2d(use_anisotropic=True, epsilon_Sigma=0.2)

print("\nIsotropic:")
print(history_iso.summary())
print("\nAnisotropic:")
print(history_aniso.summary())
```

```python
# Visualize isotropic (flat manifold - should show minimal z-variation)
from fragile.experiments.manifold import ManifoldExplorer

explorer_iso = ManifoldExplorer(
    history=history_iso,
    fitness_op=fitness_op,
    epsilon_Sigma=0.2,
    curvature_scale=2.0,
    show_manifold=True,
    show_principal_directions=False,  # Less clutter
)

print("Isotropic Diffusion (should show flatter manifold):")
explorer_iso.panel()
```

```python
# Visualize anisotropic (curved manifold - should show significant z-variation)
explorer_aniso = ManifoldExplorer(
    history=history_aniso,
    fitness_op=fitness_op,
    epsilon_Sigma=0.2,
    curvature_scale=2.0,
    show_manifold=True,
    show_principal_directions=True,
)

print("Anisotropic Diffusion (should show curved manifold with varied z-heights):")
explorer_aniso.panel()
```

### Comparison Analysis

**Isotropic Diffusion:**
- Manifold should appear relatively flat (minimal z-variation)
- All directions explored equally
- No adaptation to fitness landscape structure

**Anisotropic Diffusion:**
- Manifold shows clear curvature (peaks and valleys in z)
- Principal directions show varied diffusion
- Geometry adapts to fitness landscape:
  - High curvature near optima (exploitation)
  - Low curvature in valleys (exploration)

This is the emergent geometry in action!

## Part 5: Advanced Features

### Accessing Data Programmatically

```python
# Extract geometry data for analysis
from fragile.experiments.manifold import (
    compute_riemannian_metric,
    compute_diffusion_tensor,
    compute_principal_directions,
)

# Get Hessian from history (if available) or compute on-the-fly
if history_aniso.fitness_hessians_full is not None:
    # Use recorded Hessian
    frame_idx = 50  # Mid-simulation
    hessian = history_aniso.fitness_hessians_full[frame_idx]
    positions = history_aniso.x_final[frame_idx + 1]  # +1 due to info offset
    print(f"Using recorded Hessian at frame {frame_idx}")
else:
    # Compute on-the-fly
    frame_idx = 50
    positions = history_aniso.x_final[frame_idx + 1]
    velocities = history_aniso.v_final[frame_idx + 1]
    rewards = history_aniso.rewards[frame_idx]
    alive = history_aniso.alive_mask[frame_idx]
    companions = history_aniso.companions_distance[frame_idx]
    
    hessian = fitness_op.compute_hessian(
        positions=positions,
        velocities=velocities,
        rewards=rewards,
        alive=alive,
        companions=companions,
        diagonal_only=False,
    )
    print(f"Computed Hessian on-the-fly at frame {frame_idx}")

# Compute geometry
metric = compute_riemannian_metric(hessian, epsilon_Sigma=0.2)
diffusion = compute_diffusion_tensor(metric)

print(f"\nMetric tensor shape: {metric.shape}")
print(f"Diffusion tensor shape: {diffusion.shape}")

# Analyze eigenvalues (curvature)
eigenvalues = torch.linalg.eigvalsh(metric[:10])  # First 10 walkers
print(f"\nMetric eigenvalues (first 10 walkers):\n{eigenvalues}")
print(f"\nMean eigenvalues: {eigenvalues.mean(dim=0)}")
print(f"Eigenvalue range: [{eigenvalues.min():.4f}, {eigenvalues.max():.4f}]")
```
