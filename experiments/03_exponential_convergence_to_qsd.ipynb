# Notebook 3: The Straight Line to Equilibrium

**Didactic Goal**: Provide undeniable evidence of exponential convergence to a unique Quasi-Stationary Distribution (QSD).

**Why it's Impressive**: An exponential decay curve on a log-linear scale becomes a **straight line**â€”the most iconic visualization of exponential convergence in science and engineering.

**Key Theoretical Claim**: 
$$D_{\text{KL}}(\mu_t \| \pi_{\text{QSD}}) \leq C e^{-\kappa t}$$

where:
- $\mu_t$ is the empirical distribution of the swarm at time $t$
- $\pi_{\text{QSD}}$ is the unique quasi-stationary distribution
- $\kappa > 0$ is the convergence rate
- $C$ is a constant depending on initial conditions

**What to Expect**: A straight line on the log-linear plot = exponential convergence! ðŸ“‰

## 1. Define the Target Potential and QSD

We'll use a **Mixture of Gaussians** to create a multimodal potential with a known QSD. The QSD will be the equilibrium distribution under the potential:

$$\pi_{\text{QSD}}(x) \propto \exp(-\beta U(x))$$

where $U(x)$ is our potential function.

**Note**: We now use the `create_multimodal_potential()` helper function from `fragile.experiments`.

```python
# Create multimodal potential using helper function
potential, target_mixture = create_multimodal_potential(
    dims=2,
    n_gaussians=3,
    bounds_range=(-8.0, 8.0),
    seed=42
)

# Extract parameters for visualization
centers = target_mixture.centers
stds = target_mixture.stds
weights = target_mixture.weights
dims = target_mixture.dims

print(f"âœ“ Created multimodal potential with {target_mixture.n_gaussians} modes")
print(f"  Dimensions: {dims}")
print(f"  Centers: {centers.tolist()}")
print(f"  Weights: {weights.tolist()}")
```

### Visualize the Target Potential

```python
# Create grid for visualization
x_range = np.linspace(-6, 6, 200)
y_range = np.linspace(-4, 6, 200)
X, Y = np.meshgrid(x_range, y_range)
grid_points = torch.tensor(np.stack([X.ravel(), Y.ravel()], axis=1), dtype=torch.float32)

# Evaluate potential on grid
Z_potential = potential.evaluate(grid_points).detach().numpy().reshape(X.shape)

# Compute target QSD (proportional to exp(-beta * U))
beta = 1.0  # Inverse temperature
Z_qsd = np.exp(-beta * Z_potential)
Z_qsd = Z_qsd / Z_qsd.sum()  # Normalize

# LEFT: Potential landscape with HoloViews
potential_image = hv.Image(
    (x_range, y_range, Z_potential),
    kdims=['xâ‚', 'xâ‚‚'],
    vdims='U'
).opts(
    cmap='viridis',
    colorbar=True,
    width=600,
    height=500,
    title='Potential Landscape U(x)',
    xlabel='xâ‚',
    ylabel='xâ‚‚',
    tools=['hover'],
)

# Add mode markers to potential plot
# FIX: Use size as a dimension (vdims) instead of option
mode_data_pot = pd.DataFrame({
    'x': centers[:, 0].numpy(),
    'y': centers[:, 1].numpy(),
    'size': weights.numpy() * 30
})
mode_points_pot = hv.Points(
    mode_data_pot,
    kdims=['x', 'y'],
    vdims='size',
    label='Modes'
).opts(
    size='size',
    color='red',
    marker='star',
    line_color='white',
    line_width=2,
)

potential_plot = potential_image * mode_points_pot

# RIGHT: Target QSD
qsd_image = hv.Image(
    (x_range, y_range, Z_qsd),
    kdims=['xâ‚', 'xâ‚‚'],
    vdims='Ï€_QSD'
).opts(
    cmap='plasma',
    colorbar=True,
    width=600,
    height=500,
    title='Target QSD Ï€_QSD(x) âˆ e^(-Î²U(x))',
    xlabel='xâ‚',
    ylabel='xâ‚‚',
    tools=['hover'],
)

# Add mode markers to QSD plot
mode_data_qsd = pd.DataFrame({
    'x': centers[:, 0].numpy(),
    'y': centers[:, 1].numpy(),
    'size': weights.numpy() * 30
})
mode_points_qsd = hv.Points(
    mode_data_qsd,
    kdims=['x', 'y'],
    vdims='size',
    label='Modes'
).opts(
    size='size',
    color='red',
    marker='star',
    line_color='white',
    line_width=2,
)

qsd_plot = qsd_image * mode_points_qsd

# Combine in horizontal layout
layout = (potential_plot + qsd_plot).opts(
    opts.Layout(shared_axes=False)
)

layout
```

## 2. Initialize the Geometric Gas

We'll initialize the swarm **far from equilibrium** to clearly observe the convergence process.

### âš ï¸ Important: Refactored Parameters

**Note**: This notebook uses the **refactored EuclideanGas/GeometricGas** implementation where:

1. **CloningParams now includes**:
   - Fitness computation parameters: `alpha`, `beta`, `eta`, `A`, `sigma_min`
   - Cloning decision parameters: `p_max`, `epsilon_clone`
   - `CompanionSelection` instance (replaces `epsilon_c` and `companion_selection_method`)

2. **Removed old parameters**:
   - âŒ `epsilon_c` â†’ Now in `CompanionSelection(epsilon=...)`
   - âŒ `companion_selection_method` â†’ Now `CompanionSelection(method=...)`
   - âŒ `use_inelastic_collision` â†’ Always uses inelastic collision

3. **All cloning logic** is now in `src/fragile/operators.py` with comprehensive tests

```python
# Geometric Gas parameters
N = 100  # Number of walkers
n_steps = 300  # Run for long time to observe convergence

# Import bounds and companion selection
from fragile.bounds import TorchBounds
from fragile.companion_selection import CompanionSelection
from fragile.euclidean_gas import CloningParams

# Define bounds to match visualization range
bounds = TorchBounds(
    low=torch.tensor([-6.0, -6.0]),
    high=torch.tensor([6.0, 6.0])
)
lambda_alg = 1.0
# Create companion selection strategy
companion_selection = CompanionSelection(
    method="uniform",      # Uniform random selection (simplest, fastest)
    lambda_alg=lambda_alg,        # Position-only distance (no velocity similarity)
)

# Create parameters with refactored CloningParams
params = GeometricGasParams(
    # ========================================================================
    # BASIC PARAMETERS
    # ========================================================================
    N=N,         # Number of walkers in the swarm
    d=dims,      # Spatial dimension (d=2 for this experiment)
    potential=potential,  # Target potential U(x) defining the QSD
    
    # ========================================================================
    # LANGEVIN DYNAMICS PARAMETERS (Kinetic Operator)
    # ========================================================================
    langevin=LangevinParams(
        gamma=1.0,      # Friction coefficient Î³ (controls damping of velocities)
        beta=1.0,       # Inverse temperature Î² = 1/T (controls thermal noise)
        delta_t=0.05    # Time step size Î”t for BAOAB integrator
    ),
    
    # ========================================================================
    # CLONING OPERATOR PARAMETERS (Refactored - operators.py)
    # ========================================================================
    cloning=CloningParams(
        sigma_x=0.1,           # Position jitter scale Ïƒ_x when cloning
        lambda_alg=lambda_alg,        # Velocity weight in algorithmic distance
        alpha_restitution=0.9, # Coefficient of restitution Î± âˆˆ [0,1]
        
        # Fitness computation parameters (from operators.py)
        alpha=1.0,             # Reward channel exponent in fitness
        beta=1.0,              # Diversity channel exponent in fitness
        eta=0.1,               # Positivity floor parameter
        A=2.0,                 # Upper bound for logistic rescale
        sigma_min=1e-8,        # Regularization for patched standardization
        
        # Cloning decision parameters (from operators.py)
        p_max=1.0,             # Maximum cloning probability threshold
        epsilon_clone=0.01,    # Regularization for cloning score
        
        # Companion selection (replaces old epsilon_c + method)
        companion_selection=companion_selection,
    ),
    
    # ========================================================================
    # LOCALIZATION KERNEL PARAMETERS (Geometric Gas Extension)
    # ========================================================================
    localization=LocalizationKernelParams(
        rho=100,               # Localization scale Ï
        kernel_type="gaussian" # Type of localization kernel
    ),
    
    # ========================================================================
    # ADAPTIVE MECHANISMS PARAMETERS (Geometric Gas Extension)
    # ========================================================================
    adaptive=AdaptiveParams(
        epsilon_F=0.05,        # Adaptation rate Îµ_F for fitness-driven force
        nu=0.02,               # Viscous coupling coefficient Î½
        epsilon_Sigma=0.01,    # Hessian regularization parameter Îµ_Î£
        rescale_amplitude=1.0, # Amplitude A of rescale function g_A
        sigma_var_min=0.1,     # Minimum variance for Z-score regularization
        viscous_length_scale=100  # Length scale for viscous coupling
    ),
    
    # ========================================================================
    # BOUNDARY AND DEVICE SETTINGS
    # ========================================================================
    bounds=bounds,       # Position bounds
    device="cpu",        # PyTorch device: "cpu" or "cuda"
    dtype="float32",     # PyTorch dtype: "float32" or "float64"
    freeze_best=False    # Whether to freeze highest-fitness walker
)

# Measurement function: negative potential (higher is better)
def measurement_fn(x):
    """Measurement function d: X â†’ â„ for fitness evaluation.
    
    We use d(x) = -U(x) so that:
    - Lower potential U(x) â†’ higher fitness d(x)
    - Walkers are attracted to low-potential regions (modes of QSD)
    """
    return -potential.evaluate(x)

# Create Geometric Gas instance
gas = GeometricGas(params, measurement_fn=measurement_fn)

# Initialize swarm FAR FROM EQUILIBRIUM
# Start uniformly in a corner away from all modes
x_init = torch.rand(N, dims) * 2.0 + 4.0  # In corner [4,6] x [4,6] (within bounds!)
v_init = torch.randn(N, dims) * 0.1       # Small initial velocities

print(f"âœ“ Initialized Geometric Gas with REFACTORED PARAMETERS")
print(f"  N walkers: {N}")
print(f"  Dimensions: {dims}")
print(f"  Bounds: [{bounds.low.tolist()}, {bounds.high.tolist()}]")
print(f"  Time steps: {n_steps}")
print(f"\nðŸ“‹ Key Parameter Settings:")
print(f"  Langevin: Î³={params.langevin.gamma}, Î²={params.langevin.beta}, Î”t={params.langevin.delta_t}")
print(f"  Cloning: Ïƒ_x={params.cloning.sigma_x}, Î±_rest={params.cloning.alpha_restitution}")
print(f"  Fitness: alpha={params.cloning.alpha}, beta={params.cloning.beta}, Î·={params.cloning.eta}")
print(f"  Companion Selection: {params.cloning.companion_selection.method}")
print(f"  Localization: Ï={params.localization.rho}")
print(f"  Adaptive: Îµ_F={params.adaptive.epsilon_F}, Î½={params.adaptive.nu}")
print(f"\n  Initial position: [{x_init.min():.2f}, {x_init.max():.2f}] (within bounds)")
print(f"  Dead walkers will resurrect via cloning!")
```

## 3. Run the Simulation and Track Convergence

We use the `ConvergenceExperiment` class to run the simulation and automatically track:
1. **KL-divergence** from the target QSD (approximate using KDE)
2. **Wasserstein-2 distance** as an alternative metric
3. **Lyapunov function** $V_{\text{total}}$

**Note**: All computational logic is now in `fragile.experiments.convergence_analysis`.

```python
# Create convergence analyzer
analyzer = ConvergenceAnalyzer(
    target_mixture=target_mixture,
    target_centers=centers,
    target_weights=weights
)

# Create experiment with snapshot times
snapshot_times = [0, 6, 16, 50, 100, n_steps]
experiment = ConvergenceExperiment(
    gas=gas,
    analyzer=analyzer,
    save_snapshots_at=snapshot_times
)

# Run the experiment!
print("Running simulation...")
print("This may take a few minutes for long trajectories.\n")

metrics, snapshots = experiment.run(
    n_steps=n_steps,
    x_init=x_init,
    v_init=v_init,
    measure_every=2,
    verbose=True
)

print("\nâœ“ Simulation complete!")
print(f"  Total measurements: {len(metrics.time)}")
print(f"  Snapshots saved: {len(snapshots)}")
```

## 4. The Main Result: Exponential Convergence (The Straight Line!)

### Plot KL-Divergence on Log Scale

**Key Observation**: If convergence is exponential with rate $\kappa$, then:
$$\log(D_{\text{KL}}(t)) \approx \log(C) - \kappa t$$

This is a **straight line** on a semi-log plot!

```python
# Get valid KL-divergence data (filters out inf/nan)
time_valid, kl_valid = metrics.get_valid_kl_data()

# Fit exponential decay
kl_fit = metrics.fit_exponential_decay("kl_divergence", fit_start_time=100)

# LEFT: Log-scale KL divergence (THE STRAIGHT LINE!)
kl_data = pd.DataFrame({'time': time_valid, 'KL': kl_valid})
kl_curve = hv.Curve(
    kl_data,
    kdims='time',
    vdims='KL',
    label='KL Divergence'
).opts(
    color='blue',
    line_width=2,
    alpha=0.7,
    logy=True,
    width=700,
    height=500,
    xlabel='Time (steps)',
    ylabel='KL Divergence D_KL(Î¼_t || Ï€_QSD)',
    title='ðŸŽ¯ Exponential Convergence: The Straight Line!',
    fontscale=1.2,
    tools=['hover'],
    show_grid=True,
)

# Add exponential fit if available
if kl_fit is not None:
    kappa, C = kl_fit
    time_fit = time_valid[time_valid >= 1]
    kl_fitted = C * np.exp(-kappa * time_fit)
    
    fit_data = pd.DataFrame({'time': time_fit, 'fit': kl_fitted})
    fit_curve = hv.Curve(
        fit_data,
        kdims='time',
        vdims='fit',
        label=f'Exponential Fit: C e^(-Îºt)\nÎº = {kappa:.4f}'
    ).opts(
        color='red',
        line_width=2,
        line_dash='dashed',
        logy=True,
    )
    
    kl_plot = kl_curve * fit_curve
    
    print(f"\nðŸ“Š Fitted Convergence Rate: Îº = {kappa:.4f}")
    print(f"   Half-life: t_1/2 = {np.log(2)/kappa:.2f} steps")
else:
    kl_plot = kl_curve

# RIGHT: Wasserstein distance (alternative metric)
time_arr = np.array(metrics.time)
w2_arr = np.array(metrics.wasserstein_distance)

w2_data = pd.DataFrame({'time': time_arr, 'W2': w2_arr})
w2_curve = hv.Curve(
    w2_data,
    kdims='time',
    vdims='W2',
    label='Wasserstein-2 Distance'
).opts(
    color='green',
    line_width=2,
    alpha=0.7,
    logy=True,
    width=700,
    height=500,
    xlabel='Time (steps)',
    ylabel='Wâ‚‚(Î¼_t, Ï€_QSD)',
    title='Wasserstein Distance to QSD',
    fontscale=1.2,
    tools=['hover'],
    show_grid=True,
)

# Combine in horizontal layout
convergence_plots = (kl_plot + w2_curve).opts(
    opts.Layout(shared_axes=False)
)

convergence_plots
```

## 5. Lyapunov Function Decay

The **Lyapunov function** $V_{\text{total}}$ should also decay exponentially.

```python
time_arr = np.array(metrics.time)
lyap_arr = np.array(metrics.lyapunov_value)

# Plot Lyapunov function on log scale
lyap_data = pd.DataFrame({'time': time_arr, 'V_total': lyap_arr})
lyap_curve = hv.Curve(
    lyap_data,
    kdims='time',
    vdims='V_total',
    label='V_total(t)'
).opts(
    color='purple',
    line_width=2,
    alpha=0.7,
    logy=True,
    width=900,
    height=500,
    xlabel='Time (steps)',
    ylabel='V_total(t)',
    title='Lyapunov Function: Exponential Decay',
    fontscale=1.2,
    tools=['hover'],
    show_grid=True,
)

# Fit exponential decay
lyap_fit = metrics.fit_exponential_decay("lyapunov_value", fit_start_time=100)
if lyap_fit is not None:
    kappa_lyap, C_lyap = lyap_fit
    time_fit = time_arr[time_arr >= 100]
    lyap_fitted = C_lyap * np.exp(-kappa_lyap * time_fit)
    
    fit_data = pd.DataFrame({'time': time_fit, 'fit': lyap_fitted})
    fit_curve = hv.Curve(
        fit_data,
        kdims='time',
        vdims='fit',
        label=f'Exponential Fit: Îº_lyap = {kappa_lyap:.4f}'
    ).opts(
        color='red',
        line_width=2,
        line_dash='dashed',
        logy=True,
    )
    
    lyap_plot = lyap_curve * fit_curve
    
    print(f"\nðŸ“Š Lyapunov Function Decay Rate: Îº_lyap = {kappa_lyap:.4f}")
else:
    lyap_plot = lyap_curve

print("\nâœ¨ The Lyapunov function also shows exponential decay, confirming stability!")

lyap_plot
```

## 5. Interactive Swarm Evolution

**Interactive Visualization**: Use the time slider to watch the swarm evolve from initial chaos to the target QSD!

```python
# Create interactive evolution plot using HoloMap with time slider
from holoviews import HoloMap

def create_evolution_frame(time_idx):
    """Create a single frame for the given time index."""
    positions = snapshots[time_idx].detach().numpy()
    
    # Background: target QSD
    qsd_bg = hv.Image(
        (x_range, y_range, Z_qsd),
        kdims=['xâ‚', 'xâ‚‚'],
        vdims='density'
    ).opts(
        cmap='Greys',
        alpha=0.3,
        colorbar=False,
    )
    
    # Swarm positions
    swarm_points = hv.Points(
        positions,
        kdims=['xâ‚', 'xâ‚‚']
    ).opts(
        size=7,
        color='blue',
        alpha=0.6,
        line_color='black',
        line_width=0.5,
    )
    
    # Mode markers - FIX: Use constant size to avoid Bokeh error
    mode_data = pd.DataFrame({
        'x': centers[:, 0].numpy(),
        'y': centers[:, 1].numpy(),
        'size': weights.numpy() * 30
    })
    mode_markers = hv.Points(
        mode_data,
        kdims=['x', 'y'],
        vdims='size',
        label='Modes'
    ).opts(
        size='size',
        color='red',
        marker='star',
        line_color='white',
        line_width=2,
    )
    
    # Get convergence metrics at this time
    # Find closest metric time
    metric_times = np.array(metrics.time)
    closest_idx = np.argmin(np.abs(metric_times - time_idx))
    kl_value = metrics.kl_divergence[closest_idx] if closest_idx < len(metrics.kl_divergence) else float('nan')
    w2_value = metrics.wasserstein_distance[closest_idx] if closest_idx < len(metrics.wasserstein_distance) else float('nan')
    
    # Combine layers
    plot = (qsd_bg * swarm_points * mode_markers).opts(
        xlim=(-6, 6),
        ylim=(-4, 6),
        xlabel='xâ‚',
        ylabel='xâ‚‚',
        title=f'Time t = {time_idx} | KL = {kl_value:.4f} | Wâ‚‚ = {w2_value:.4f}',
        width=800,
        height=700,
        show_grid=True,
        fontscale=1.2,
    )
    
    return plot

# Create HoloMap with time slider
time_indices = sorted(snapshots.keys())
evolution_holomap = HoloMap(
    {t: create_evolution_frame(t) for t in time_indices},
    kdims='time'
)

# Configure the slider widget
evolution_interactive = evolution_holomap.opts(
    # opts.HoloMap(
    #     widget_type='scrubber',  # Interactive slider
    #     widget_location='bottom',
    # )
)

print(f"âœ¨ Interactive evolution plot created!")
print(f"   Use the slider to explore {len(time_indices)} snapshots from t=0 to t={max(time_indices)}")
print(f"   Watch the swarm migrate from the initial corner to cover all three modes!")

evolution_interactive
```

```python
# Create enhanced interactive dashboard with Panel
import panel.widgets as pnw

# Time slider widget
time_slider = pnw.DiscreteSlider(
    name='Time',
    options=time_indices,
    value=0,
    width=700
)

# Play button for animation
play_button = pnw.Player(
    name='Animation',
    start=0,
    end=len(time_indices) - 1,
    value=0,
    interval=500,  # milliseconds between frames
    width=700,
    show_loop_controls=True,
)

# Link play button to actual time values
@pn.depends(play_button.param.value)
def get_time_from_player(player_idx):
    return time_indices[player_idx]

# Create synchronized plot function
@pn.depends(time_slider.param.value, play_button.param.value)
def create_synced_plot(slider_time, player_idx):
    """Create plot synchronized with both slider and player."""
    # Prefer slider if it's been moved, otherwise use player
    time_idx = slider_time if slider_time != time_indices[0] else time_indices[player_idx]
    
    positions = snapshots[time_idx].detach().numpy()
    
    # Background: target QSD
    qsd_bg = hv.Image(
        (x_range, y_range, Z_qsd),
        kdims=['xâ‚', 'xâ‚‚'],
        vdims='density'
    ).opts(
        cmap='Greys',
        alpha=0.3,
        colorbar=False,
    )
    
    # Swarm positions with color gradient based on distance from center
    swarm_points = hv.Points(
        positions,
        kdims=['xâ‚', 'xâ‚‚']
    ).opts(
        size=8,
        color='blue',
        alpha=0.7,
        line_color='black',
        line_width=0.5,
    )
    
    # Mode markers - FIX: Use size as dimension
    mode_data = pd.DataFrame({
        'x': centers[:, 0].numpy(),
        'y': centers[:, 1].numpy(),
        'size': weights.numpy() * 30
    })
    mode_markers = hv.Points(
        mode_data,
        kdims=['x', 'y'],
        vdims='size',
        label='Modes'
    ).opts(
        size='size',
        color='red',
        marker='star',
        line_color='white',
        line_width=2,
    )
    
    # Get metrics
    metric_times = np.array(metrics.time)
    closest_idx = np.argmin(np.abs(metric_times - time_idx))
    kl_value = metrics.kl_divergence[closest_idx] if closest_idx < len(metrics.kl_divergence) else float('nan')
    w2_value = metrics.wasserstein_distance[closest_idx] if closest_idx < len(metrics.wasserstein_distance) else float('nan')
    lyap_value = metrics.lyapunov_value[closest_idx] if closest_idx < len(metrics.lyapunov_value) else float('nan')
    
    # Combine
    plot = (qsd_bg * swarm_points * mode_markers).opts(
        xlim=(-6, 6),
        ylim=(-4, 6),
        xlabel='xâ‚',
        ylabel='xâ‚‚',
        title=f'Swarm Evolution at t = {time_idx}',
        width=700,
        height=650,
        show_grid=True,
        fontscale=1.1,
    )
    
    return plot

# Create metrics display
@pn.depends(time_slider.param.value, play_button.param.value)
def create_metrics_display(slider_time, player_idx):
    """Display current metrics."""
    time_idx = slider_time if slider_time != time_indices[0] else time_indices[player_idx]
    
    metric_times = np.array(metrics.time)
    closest_idx = np.argmin(np.abs(metric_times - time_idx))
    
    kl_value = metrics.kl_divergence[closest_idx] if closest_idx < len(metrics.kl_divergence) else float('nan')
    w2_value = metrics.wasserstein_distance[closest_idx] if closest_idx < len(metrics.wasserstein_distance) else float('nan')
    lyap_value = metrics.lyapunov_value[closest_idx] if closest_idx < len(metrics.lyapunov_value) else float('nan')
    
    metrics_text = f"""
    ## Convergence Metrics at t = {time_idx}
    
    **KL-Divergence**: `{kl_value:.6f}`  
    **Wasserstein-2 Distance**: `{w2_value:.6f}`  
    **Lyapunov Function**: `{lyap_value:.6f}`  
    
    ---
    
    ### Progress
    Time: **{time_idx} / {max(time_indices)}** ({100*time_idx/max(time_indices):.1f}%)
    """
    
    return pn.pane.Markdown(metrics_text, width=300)

# Create mini convergence plot that updates with time
@pn.depends(time_slider.param.value, play_button.param.value)
def create_mini_convergence_plot(slider_time, player_idx):
    """Show convergence up to current time."""
    time_idx = slider_time if slider_time != time_indices[0] else time_indices[player_idx]
    
    # Filter data up to current time
    time_arr = np.array(metrics.time)
    mask = time_arr <= time_idx
    
    time_subset = time_arr[mask]
    kl_subset = np.array(metrics.kl_divergence)[mask]
    
    # Remove invalid values
    valid_mask = ~(np.isnan(kl_subset) | np.isinf(kl_subset))
    time_valid_subset = time_subset[valid_mask]
    kl_valid_subset = kl_subset[valid_mask]
    
    if len(kl_valid_subset) > 0:
        kl_data = pd.DataFrame({'time': time_valid_subset, 'KL': kl_valid_subset})
        kl_curve = hv.Curve(
            kl_data,
            kdims='time',
            vdims='KL'
        ).opts(
            color='blue',
            line_width=2,
            logy=True,
            width=300,
            height=200,
            xlabel='Time',
            ylabel='KL Divergence',
            title='Convergence Progress',
            show_grid=True,
        )
        
        # Add current time marker
        if time_idx > 0:
            current_kl = kl_valid_subset[-1] if len(kl_valid_subset) > 0 else None
            if current_kl is not None:
                marker = hv.Scatter(
                    [(time_idx, current_kl)],
                    kdims='time',
                    vdims='KL'
                ).opts(
                    size=10,
                    color='red',
                    marker='o',
                )
                return kl_curve * marker
        
        return kl_curve
    else:
        # Empty plot
        return hv.Curve([]).opts(width=300, height=200, title='Convergence Progress')

# Assemble dashboard
dashboard = pn.Column(
    pn.pane.Markdown("# ðŸŽ¬ Interactive Swarm Evolution Dashboard", width=800),
    pn.pane.Markdown("Use the **player** to watch the animation or **slider** for manual control.", width=800),
    play_button,
    time_slider,
    pn.Row(
        create_synced_plot,
        pn.Column(
            create_metrics_display,
            create_mini_convergence_plot,
        )
    ),
)

print("âœ¨ Enhanced interactive dashboard created!")
print("   â€¢ Use PLAY button for automatic animation")
print("   â€¢ Use TIME SLIDER for manual control")
print("   â€¢ Watch metrics update in real-time!")

dashboard
```

## 6. Marginal Distributions: Empirical vs Target

```python
# Final swarm positions
final_time = max(snapshots.keys())
final_positions = snapshots[final_time].detach().numpy()

# Compute empirical statistics
emp_mean = final_positions.mean(axis=0)
emp_cov = np.cov(final_positions.T)

# Target statistics
target_mean = (centers * weights.unsqueeze(1)).sum(dim=0).numpy()
target_cov_approx = sum(
    weights[i].item() * (np.outer(centers[i].numpy(), centers[i].numpy()) + np.diag(stds[i].numpy()**2))
    for i in range(len(weights))
) - np.outer(target_mean, target_mean)

print("\nðŸ“Š Statistical Comparison (Final vs Target):\n")
print(f"Empirical Mean:   {emp_mean}")
print(f"Target Mean:      {target_mean}")
print(f"Mean Error:       {np.linalg.norm(emp_mean - target_mean):.4f}\n")

print(f"Empirical Covariance:\n{emp_cov}\n")
print(f"Target Covariance (approx):\n{target_cov_approx}\n")
print(f"Covariance Frobenius Error: {np.linalg.norm(emp_cov - target_cov_approx, 'fro'):.4f}")

# Visual comparison of marginal distributions using HoloViews
marginal_plots = []

for dim_idx in range(dims):
    # Empirical histogram
    hist_data = final_positions[:, dim_idx]
    frequencies, edges = np.histogram(hist_data, bins=30, density=True)
    hist_hv = hv.Histogram((edges, frequencies), label='Empirical (Final Swarm)').opts(
        fill_color='blue',
        alpha=0.6,
        line_color='black',
    )
    
    # Target density (mixture of 1D Gaussians)
    x_range_1d = np.linspace(-6, 6, 200)
    target_density = sum(
        weights[i].item() / (np.sqrt(2 * np.pi) * stds[i, dim_idx].item()) *
        np.exp(-0.5 * ((x_range_1d - centers[i, dim_idx].item()) / stds[i, dim_idx].item())**2)
        for i in range(len(weights))
    )
    
    target_curve = hv.Curve(
        (x_range_1d, target_density),
        kdims='x',
        vdims='density',
        label='Target QSD'
    ).opts(
        color='red',
        line_width=2,
    )
    
    # Combine
    marginal_plot = (hist_hv * target_curve).opts(
        width=600,
        height=400,
        xlabel=f'x_{dim_idx+1}',
        ylabel='Density',
        title=f'Marginal Distribution: Dimension {dim_idx+1}',
        show_grid=True,
        fontscale=1.1,
        legend_position='top_right',
    )
    
    marginal_plots.append(marginal_plot)

# Layout horizontally
marginal_layout = hv.Layout(marginal_plots).cols(2)

print("\nâœ¨ The empirical distribution closely matches the target QSD!")

marginal_layout
```
